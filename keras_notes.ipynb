```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow 
from tensorflow.python import keras
from keras.models import Sequential 
from keras.layers import Activation, Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

# Train

```python
train_labels = []
train_samples = []
```

#Experimental data:
An experiment drug was tested on individuals from ages 13 to 100
The trial had 2100 participants. Half were under 65 years old, half were more than 65 years old.
95% of patients 65 or older experienced side effects
95% of patients under 65 experienced no side effects

```python

for i in range(1000):
    random_younger = np.random.randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = np.random.randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range(50):
    random_younger = np.random.randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = np.random.randint(65, 100)
    train_samples.append(random_older)
    train_labels.append(0)
    
```

```python
train_labels = np.asarray(train_labels)
train_samples = np.asarray(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_data_samples:
    print(i)
```

# Make model

```python
model = Sequential([
    Dense(16,input_shape=(1,),activation="relu"),
    Dense(32, activation="relu"),
    Dense(2, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

valid_set = [(sample, label), (sample, label), (sample, label),... (sample, label), (sample, label)]
ans pass this array in mode.fit() function as:
model.fit(scaled_data_samples, train_labels, validation_data = valid_set, batch_size=10, epochs=20, shuffle=True, verbose=2)

So, model will be trained on the scaled_data_samples and train_labels and will test on valid_set.

Other way of doing this is by passing an argument in the model.fit() function
model.fit(scaled_data_samples, train_labels, validation_split = 0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)

```python
model.fit(scaled_data_samples, train_labels, validation_split = 0.2, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

# Predict

```python
test_samples = []
test_labels = []

for i in range(10):
    #5% of the younger individuals who did experience the side effects
    random_younger = np.random.randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #5% of the older individuals who did not experience the side effects
    random_older = np.random.randint(65, 100)
    test_samples.append(random_older)
    test_labels.append(0)
    
for i in range(200):
    #95% of the younger individuals who did not experience the side effects
    random_younger = np.random.randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #95% of the older individuals who did experience the side effects
    random_older = np.random.randint(65, 100)
    test_samples.append(random_older)
    test_labels.append(0)
    
    
    
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)


scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
predictions= model.predict(scaled_test_samples, batch_size = 10, verbose=0)
```

```python
for i in predictions:
    print(i)
```

```python
rounded_predictions= model.predict_classes(scaled_test_samples, batch_size = 10, verbose=0)
```

```python
for i in rounded_predictions:
    print(i)
```

# Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(test_labels, rounded_predictions)
```

```python
# copied from matplotlib website

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'has_side_effects']
plot_confusion_matrix(cm, cm_plot_labels, title='confusion_matrix')
```
